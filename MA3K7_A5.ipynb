```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sum_of_ring(i):
    # Solve for n
    n = np.sqrt(i / 4)
    
    # Round n to the nearest integer (up in case of decimal)
    n_rounded = np.ceil(n)
    
    #  Calculate k_n and l_n
    k_n = 4 * n_rounded**2 - 8 * n_rounded + 5
    l_n = 4 * n_rounded**2

    # Calculate the sum
    number_of_terms = l_n - k_n + 1
    total_sum = (k_n + l_n) * number_of_terms // 2 

    return total_sum
```

```python
# Test function with input 512 
print("The sum of the the ring with 512 is", sum_of_ring(512))

# Define the range of i values
i_values = np.arange(1, 2001)

# Calculate the sum for each i value
sum_values = [sum_of_ring(i) for i in i_values]

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(i_values, sum_values, label='Sum of Ring')
plt.xlabel('i value')
plt.ylabel('Sum of Ring')
plt.title('Sum of Ring for values of i from 1 to 2000')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the percentage of times the sum changes
change_counts = np.count_nonzero(np.diff(sum_values) != 0)
percentage_changes = (change_counts / len(i_values)) * 100
```
