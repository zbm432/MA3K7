```python
# Import necessary libraries 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Define probability function 
def p(n):
    if n < 2:
        raise ValueError('number must be greater than 1')
    elif n == 2:
        return 0.5
    elif n == 3:
        return 0.75 
    else:
        # Use recursive formula 
        return 1 - 0.5 * p(n - 1)
```

```python
# Set up numbers between 2 and 49
numbers = range(2, 50)

# Iterate through all n and caclulate probability 
probabilities = [p(n) for n in numbers]

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(numbers, probabilities, marker='o', label='p(n)')
plt.axhline(y=2/3, color='r', linestyle='-', label='y = 2/3')
plt.title('Probability of each number')
plt.xlabel('n')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```
